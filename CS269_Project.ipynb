```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sc
import math
from scipy.special import erf
from scipy.special import erfc

e = math.exp(1)
pi = math.pi

def G2_1(q1,q2,lamb,numSimuls):
    x1 = lamb*q1
    x2 = lamb*q2
    res = 0
    for i in range(numSimuls):
        a = np.random.normal()
        b = np.random.normal()
        c = np.random.normal()
        d = np.random.normal()

        Ap = a**2 + c**2
        A = Ap + x1
        Bp = b**2 + d**2
        B = Bp + x2
        C = a*b + c*d
        det = A*B - C**2
        temp = (B**2 + C**2)/det**2
        res += temp
    res /= numSimuls
    res *= lamb**2
    return res

def G2_2(q1,q2,lamb,numSimuls):
    x1 = lamb*q1
    x2 = lamb*q2
    res = 0
    for i in range(numSimuls):
        a = np.random.normal()
        b = np.random.normal()
        c = np.random.normal()
        d = np.random.normal()

        Ap = a**2 + c**2
        A = Ap + x1
        Bp = b**2 + d**2
        B = Bp + x2
        C = a*b + c*d
        det = A*B - C**2
        temp = (A**2 + C**2)/det**2
        res += temp
    res /= numSimuls
    res *= lamb**2
    return res

def G3_1(q1,q2,lamb,numSimuls):
    x1 = lamb*q1
    x2 = lamb*q2
    res = 0
    for i in range(numSimuls):
        a = np.random.normal()
        b = np.random.normal()
        c = np.random.normal()
        d = np.random.normal()
        e = np.random.normal()
        f = np.random.normal()

        Ap = a**2 + c**2 + e**2
        A = Ap + x1
        Bp = b**2 + d**2 + f**2
        B = Bp + x2
        C = a*b + c*d
        det = A*B - C**2
        det = (A*B - C**2)
        temp = (B**2 + C**2)/det**2
        res += temp
    res /= numSimuls
    res *= lamb**2
    return res    

def G3_1_ub_temp(q1,q2,lamb,numSimuls):
    x1 = lamb*q1
    x2 = lamb*q2
    res = 0
    for i in range(numSimuls):
        a = np.random.normal()
        b = np.random.normal()
        c = np.random.normal()
        d = np.random.normal()
        e = np.random.normal()
        f = np.random.normal()

        Ap = a**2 + c**2 + e**2
        A = Ap + x1
        Bp = b**2 + d**2 + f**2
        B = Bp + x2
        C = a*b + c*d
        det = A*B - C**2
        det = (A*B - C**2) -(Ap*Bp - C**2)
        temp = (B**2 + C**2)/det**2
        res += temp
    res /= numSimuls
    res *= lamb**2
    return res

def G3_1_ub_temp2(q1,q2,lamb,numSimuls):
    x1 = lamb*q1
    x2 = lamb*q2
    res = 0
    for i in range(numSimuls):
        a = np.random.normal()
        b = np.random.normal()
        c = np.random.normal()
        d = np.random.normal()
        e = np.random.normal()
        f = np.random.normal()

        Ap = a**2 + c**2 + e**2
        A = Ap + x1
        Bp = b**2 + d**2 + f**2
        B = Bp + x2
        C = a*b + c*d

        temp1 = B**2 + 3*(a**2)*(b**2)
        temp2 = 1/(Ap*lamb*q2 + Bp*lamb*q1 + q1*q2*lamb**2)**2
        temp = temp1 * temp2

        res += temp
    res /= numSimuls
    res *= lamb**2
    return res

def G3_2(q1,q2,lamb,numSimuls):
    x1 = lamb*q1
    x2 = lamb*q2
    res = 0
    for i in range(numSimuls):
        a = np.random.normal()
        b = np.random.normal()
        c = np.random.normal()
        d = np.random.normal()
        e = np.random.normal()
        f = np.random.normal()

        Ap = a**2 + c**2 + e**2
        A = Ap + x1
        Bp = b**2 + d**2 + f**2
        B = Bp + x2
        C = a*b + c*d
        det = A*B - C**2
        det = (A*B - C**2)
        temp = (A**2 + C**2)/det**2
        res += temp
    res /= numSimuls
    res *= lamb**2
    return res    

def G3_2_ub_temp(q1,q2,lamb,numSimuls):
    x1 = lamb*q1
    x2 = lamb*q2
    res = 0
    for i in range(numSimuls):
        a = np.random.normal()
        b = np.random.normal()
        c = np.random.normal()
        d = np.random.normal()
        e = np.random.normal()
        f = np.random.normal()

        Ap = a**2 + c**2 + e**2
        A = Ap + x1
        Bp = b**2 + d**2 + f**2
        B = Bp + x2
        C = a*b + c*d
        det = A*B - C**2
        det = (A*B - C**2) - (Ap*Bp - C**2)
        temp = (A**2 + C**2)/det**2
        res += temp
    res /= numSimuls
    res *= lamb**2
    return res  

def G3_2_ub_temp2(q1,q2,lamb,numSimuls):
    x1 = lamb*q1
    x2 = lamb*q2
    res = 0
    for i in range(numSimuls):
        a = np.random.normal()
        b = np.random.normal()
        c = np.random.normal()
        d = np.random.normal()
        e = np.random.normal()
        f = np.random.normal()

        Ap = a**2 + c**2 + e**2
        A = Ap + x1
        Bp = b**2 + d**2 + f**2
        B = Bp + x2
        C = a*b + c*d

        temp1 = A**2 + 3*(a**2)*(b**2)
        temp2 = 1/(Ap*lamb*q2 + Bp*lamb*q1 + q1*q2*lamb**2)**2
        temp = temp1 * temp2

        res += temp
    res /= numSimuls
    res *= lamb**2
    return res
    
def T(x):
    return 0.5 * sc.expn(1,x/2) * e**(x/2)

def S(x):
    return 0.5 * sc.expn(2,x/2) * e**(x/2)

def h(x):
    res1 = (0.5 * pi * x * e**x)**0.5
    res2 = erfc((x/2)**0.5)
    res = 1 - res1 * res2
    return res

def G2_1_lb(q1,q2,lamb):
    x1 = lamb*q1
    x2 = lamb*q2
    res1 = (lamb/q1) * S(x1)
    res2 = (lamb**2/2) * (T(x1) - S(x1)) * (T(x2) - S(x2))
    res = res1 + res2
    return res

def G2_2_lb(q1,q2,lamb):
    x1 = lamb*q1
    x2 = lamb*q2
    res1 = (lamb/q2) * S(x2)
    res2 = (lamb**2/2) * (T(x1) - S(x1)) * (T(x2) - S(x2))
    res = res1 + res2
    return res

def G3_1_ub(q1,q2,lamb):
    return ((3+lamb*q2)*h(lamb*q1)/(q1 * q2)) + (1 - lamb*q1*h(lamb*q1))*(1 - lamb*q2*h(lamb*q2))/(3*q1*q2) 

def G3_2_ub(q1,q2,lamb):
    return ((3+lamb*q1)*h(lamb*q2)/(q1 * q2)) + (1 - lamb*q1*h(lamb*q1))*(1 - lamb*q2*h(lamb*q2))/(3*q1*q2) 
```

```python
q1 = 1
q2 = 0.5
numSimuls = 300
arr1 = []
arr2 = []
arr1a = []
arr2a = []
for lamb in range(1,100):
    x1 = lamb*q1
    x2 = lamb*q2
    arr1.append(G2_1(q1,q2,lamb,numSimuls) - G3_1(q1,q2,lamb,numSimuls))
    arr1a.append(G2_1_lb(q1,q2,lamb) - G3_1_ub(q1,q2,lamb))
    # arr1a.append(G2_1_lb(q1,q2,lamb) - G3_1_ub_temp2(q1,q2,lamb,numSimuls))

    arr2.append(G2_2(q1,q2,lamb,numSimuls) - G3_2(q1,q2,lamb,numSimuls))
    arr2a.append(G2_2_lb(q1,q2,lamb) - G3_2_ub(q1,q2,lamb))
    # arr2a.append(G2_2_lb(q1,q2,lamb) - G3_2_ub_temp2(q1,q2,lamb,numSimuls))

lower = 0
plt.plot(arr1[lower:],label='alpha1 - beta1')
plt.plot(arr1a[lower:],label='alpha1 - beta1 (lb)')
plt.plot(arr2[lower:],label='alpha2 - beta2')
plt.plot(arr2a[lower:],label='alpha2 - beta2 (lb)')
plt.xlabel('lambda')
plt.title('Lower Bounds on alpha - beta (Weak)')

print(arr1[-1])
print(arr1a[-1])
print(arr2[-1])
print(arr2a[-1])

plt.legend()
plt.show()
```

```python
def G1(q,lamb,numSimuls):
    res = 0
    for i in range(numSimuls):
        a = np.random.normal()
        res += 1/(a**2 + lamb*q)**2
    res /= numSimuls
    res *= lamb**2
    return res

def G2(q,lamb,numSimuls):
    res = 0
    for i in range(numSimuls):
        a = np.random.normal()
        b = np.random.normal()
        res += 1/(a**2 + b**2 + lamb*q)**2
    res /= numSimuls
    res *= lamb**2
    return res

def H1(q,lamb,numSimuls):
    res = 0
    for i in range(numSimuls):
        a = np.random.normal()
        res += a**2/(a**2 + lamb*q)**2
    res /= numSimuls
    return res


def H2(q,lamb,numSimuls):
    res = 0
    for i in range(numSimuls):
        a = np.random.normal()
        b = np.random.normal()
        res += (a**2 + b**2)/(a**2 + b**2 + lamb*q)**2
    res /= numSimuls
    return res
```
